```python
import pandas as pd
import statsmodels.api as sm
```

```python
df = pd.read_csv('../data/coffeeleaflet.csv')
df.head()
```

```python
df= df.loc[:, ['treatment', 'awarewaste']]
df = df.dropna()
D = df['treatment']
D = sm.add_constant(D) # add intercept
Y = df['awarewaste']
results = sm.OLS(Y, D).fit(cov_type='HC0')
print(results.summary())
```

## Intrepretation of A/B Test Results
In this example, we conduct an A/B test to evaluate the impact of a treatment (e.g., a marketing leaflet) on the awareness of waste among participants. We use the `statsmodels` library to perform an Ordinary Least Squares (OLS) regression analysis.

The treatment coefficient of 0.2801 means that being in the treatment group (receiving the leaflet) is associated with a 28.01% increase in the awareness of waste compared to the control group (not receiving the leaflet). This suggests that the leaflet had a positive effect on raising awareness about waste management among the participants.

Given an intercept of 2.01, this indicates that, on average, participants in the control group had an awareness score of 2.01. Therefore, the treatment increased this score to approximately 2.29 (2.01 + 0.2801) for those in the treatment group.
